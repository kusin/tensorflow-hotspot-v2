```python
import matplotlib.pyplot as plt
import numpy as np

# Data dari tabel
models = ['SBi-LSTM', 'SBi-LSTM-XGBoost', 'SBi-GRU', 'SBi-GRU-XGBoost']
types = ['Univariate', 'Multivariate']
metrics = ['R']

# Nilai-nilai yang diambil dari tabel (diubah ke dalam skala 0-1 untuk R)
# Data diambil secara manual dari tabel
data = {
    ('SBi-LSTM', 'Univariate'): 0.6452,
    ('SBi-LSTM', 'Multivariate'): 0.7354,
    ('SBi-LSTM-XGBoost', 'Univariate'): 0.9999,
    ('SBi-LSTM-XGBoost', 'Multivariate'): 1.0000,
    ('SBi-GRU', 'Univariate'): 0.6449,
    ('SBi-GRU', 'Multivariate'): 0.8068,
    ('SBi-GRU-XGBoost', 'Univariate'): 0.9999,
    ('SBi-GRU-XGBoost', 'Multivariate'): 1.0000
}

# Membuat barplot
bar_width = 0.35
index = np.arange(len(models))

plt.figure(figsize=(8,4))

univariate_values = [data[(model, 'Univariate')] for model in models]
multivariate_values = [data[(model, 'Multivariate')] for model in models]

bars1 = plt.bar(index, univariate_values, bar_width, label='Univariate')
bars2 = plt.bar(index + bar_width, multivariate_values, bar_width, label='Multivariate')

plt.xlabel('')
plt.ylabel('')
plt.title('Model evaluation of R value')
plt.xticks(index + bar_width/2, models)
plt.legend()

# # Menambahkan nilai R di atas setiap bar
# for bar1, bar2 in zip(bars1, bars2):
#     plt.text(bar1.get_x() + bar1.get_width()/2., bar1.get_height(), f'{bar1.get_height():.4f}', ha='center', va='bottom', fontsize=9)
#     plt.text(bar2.get_x() + bar2.get_width()/2., bar2.get_height(), f'{bar2.get_height():.4f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.grid(True)
plt.show()
```
